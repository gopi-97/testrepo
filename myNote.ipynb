### The following code checks 

```python
#code to check whether a given number is prime or ot 
num=int(input("enter a number: "))
flag=1
j=2
while(j<num):
    if(num%j==0):
        flag=0
    j+=1
if(flag):
    print("\n the number is prime:")
else:
    print("the number is not prime")
```
